```python
import pandas as pd
import numpy as np
import json
```

```python
f = open('C:/Users/91903/Downloads/data.json')
data = json.load(f)
d = data['results']
df = pd.DataFrame(d)
df = pd.concat([df.drop(['openfda'],axis = 1) , df['openfda'].apply(pd.Series,dtype=object)],axis=1)
df = df[['brand_name','adverse_reactions_table']]
df = df.dropna()
df
```

```python
arr = df.to_numpy()
len(arr)
```

```python
lists_of_tables=[]
for i in range(10):
    
    table_MN = pd.read_html(arr[i][1][0])
    print(f'Total tables: {len(table_MN)}')
    df1 = table_MN[0]
    df1.insert(loc=0, column='brand_name', value=arr[i][0][0])
    #df1['brand_name'] = arr[i][0][0]
    lists_of_tables.append(df1)
len(lists_of_tables)
```

```python
pd.concat(lists_of_tables).to_csv('openfda aadverreactions.csv')
```
